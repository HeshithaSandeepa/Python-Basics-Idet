## 1. Mean (Average) - ‡∑É‡∑è‡∂∏‡∑è‡∂±‡∑ä‚Äç‡∂∫ ‡∂Ö‡∂ú‡∂∫
### Formula:
$$\bar{x} = \frac{\sum_{i=1}^{n} x_i}{n} = \frac{x_1 + x_2 + ... + x_n}{n}$$

```python
# Ube marks 5k kiyala hithamu:
marks = [75, 82, 90, 68, 85]

# Manual way
total = sum(marks)  # Okkoma ekathu kara
count = len(marks)  # Subjects keeyada?
mean = total / count  # Beduwa

print(f"Marks: {marks}")
print(f"Total: {total}")
print(f"Subjects: {count}")
print(f"Mean (Average): {total}/{count} = {mean}")

# Numpy way
mean_numpy = np.mean(marks)
print(f"\nNumpy walin gahuwama: {mean_numpy}")
```

### Mean eka pawichchi wenne koheda?

- Average marks balanna
- Salary average eka hoyanna

### Habai podi aulak thiyenawa
Mean eka samahara welawata **outliers** nisa (e kiyanne pena hama data ekatama wada wenas loku/podi ewata) confuse wenna puluwan.

```python
salaries_normal = [30000, 35000, 32000, 28000, 33000]
print(f"salaries: {salaries_normal}")
print(f"Mean salary: Rs. {np.mean(salaries_normal):.2f}")

salaries_with_ceo = [30000, 35000, 32000, 28000, 500000]  # CEO set una
print(f"\nCEO ekka salary set eka: {salaries_with_ceo}")
print(f"Mean salary: Rs. {np.mean(salaries_with_ceo):.2f}")
print("\nDekka neda? Ara CEO ge salary eka nisa mean eka kohetada giya. Anna e wage welawata mean eka echchara set wenne na.")
```

- **Rule:** Data skewed ‡∂±‡∂∏‡∑ä ‚Üí Median use ‡∂ö‡∂ª‡∂±‡∑ä‡∂±.

```python
# Why Median? 
import seaborn as sns
import matplotlib.pyplot as plt

# Office eke Salary data (LKR) - Anthimata CEO eka
salaries = [45000, 52000, 48000, 50000, 47000, 49000, 1500000] # <- CEO outlier

print(f'Mean Salary: {np.mean(salaries):,.0f} LKR')  # Halucination
print(f'Median Salary: {np.median(salaries):,.0f} LKR') # Can Consider real value

plt.figure(figsize=(8, 3))
sns.boxplot(x=salaries, color='orange')
plt.title('Boxplot: CEO outlier rght side?')
plt.xlabel('Salary (LKR)')
plt.show()
```

## 3. Median - ‡∂∏‡∂∞‡∑ä‚Äç‡∂∫‡∑É‡∑ä‡∂Æ ‡∂Ö‡∂ú‡∂∫

Median kiyanne data tika pilowelata haduwama hariyatama **madata set wena value** eka Meka ara outliers walata ahu wenne na. Ekai meka patta useful wenne.

### Hoyanne kohomada?
1. Sort karanna oni (podi eke idan loku ekata)
2. Mada value eka ganna
   - Odd count (5, 7 wage) nam: Hariyatama mada eka
   - Even count (4, 6 wage) nam: Mada thiyena deke average eka

```python
# Odd count example ekak (Otte ganak)
data_odd = [7, 2, 9, 4, 5]  # 5k thiyenawa
sorted_odd = sorted(data_odd)
print(f"Original data: {data_odd}")
print(f"Sort karama: {sorted_odd}")
print(f"Mada index eka: {len(sorted_odd)//2}")
print(f"Median eka: {sorted_odd[len(sorted_odd)//2]}")

print("\n" + "="*50 + "\n")

# Even count example (Iratte ganak)
data_even = [7, 2, 9, 4, 5, 8]  # 6k thiyenawa
sorted_even = sorted(data_even)
mid = len(sorted_even) // 2
median_even = (sorted_even[mid-1] + sorted_even[mid]) / 2
print(f"Original data: {data_even}")
print(f"Sort karama: {sorted_even}")
print(f"Mada values deka: {sorted_even[mid-1]} saha {sorted_even[mid]}")
print(f"Median eka: ({sorted_even[mid-1]} + {sorted_even[mid]}) / 2 = {median_even}")
```

```python
# Ara CEO scene eka ayeth balamu - Median vs Mean
salaries_with_ceo = [30000, 35000, 32000, 28000, 500000]

print(f"Salaries: {salaries_with_ceo}")
print(f"Mean eka: Rs. {np.mean(salaries_with_ceo):.2f}")
print(f"Median eka: Rs. {np.median(salaries_with_ceo):.2f}")

print("\nDan harine? Median eken apita niyama thathwe balaganna puluwan. CEO ge padi loku unata median ekata aulak wune na.")
```

## 4. Mode - ‡∂¥‡∑ä‚Äç‡∂ª‡∂ö‡∑è‡∂ª ‡∂Ö‡∂ú‡∂∫

Mode kiyanne data set eke **wadiyenma repeat wena value** eka.

### Example:
Class eke kamathima color eka vote kara kiyamu:
- Blue: 5 denai
- Red: 8 denai ‚Üê **Anna eka thamai Mode eka!**
- Green: 3 denai

```python
# Mode example - Exam scores
scores = [75, 82, 90, 75, 68, 75, 85, 90, 75]

# Count karala balamu
from collections import Counter
count = Counter(scores)

print(f"Scores tika: {scores}")
print(f"Counts: {dict(count)}")
print(f"Mode eka: {stats.mode(scores)}")

print("\n75 thamai wadiyenma repeat wenne - 4 parak awilla!")
```

### üí° Insight: Mode
- **Use Case:** Categorical data (Colors, Cities) ‡∑Ä‡∂Ω Missing Values ‡∂¥‡∑î‡∂ª‡∑Ä‡∂±‡∑ä‡∂±.
- **Business:** 'Best Seller' (‡∑Ä‡∑ê‡∂©‡∑í‡∂∫‡∑ô‡∂±‡∑ä‡∂∏ repeat ‡∑Ä‡∑ô‡∂± ‡∂ë‡∂ö) ‡∑Ñ‡∑ú‡∂∫‡∂±‡∑ä‡∂±.

**Hari, dan api dannawa numbers wala 'Mada' hoyanna (Median).**

Eth hithanna oyage data set eke thiyenne numbers neme kiyala. Example ekata 'Colors' (Red, Blue, Green) hari 'City Names'. Mewage Average ekak ganna bane? 

Machine Learning waladi mewage data (Categorical Data) wala his than purawanna oni unama, api use karanne **Mode** eka. E kiyanne wadiyenma repeat wena eka dala shape wenawa.

```python
# Multiple modes thiyena welawal (Multimodal)
data_multimodal = [1, 2, 2, 3, 3, 4]
print(f"Data: {data_multimodal}")
print(f"Counts: {dict(Counter(data_multimodal))}")
try:
    mode_val = stats.multimode(data_multimodal)  # Python 3.8+
    print(f"Modes: {mode_val}")
except:
    print("Modes dekak thiyenawa machan: 2 saha 3 (nikanma dekama same waraganak awith)")
```

**Dan apita Data wala 'Center' eka (Mean, Median, Mode) gana idea ekak thiyenawa.**

Habai AI model ekakata data danna kalin, ewa adjust karanna oni (**Feature Scaling**). Samahara data godak wisirila (Spread), samahara ewa laga laga. 

Me 'Spread' eka hariyata scale karanne nathiwa model ekata dunnoth, model eka pissu kelinawa. Anna e Spread eka maninna thamai api me **Standard Deviation** eka use karanne. (Ex: StandardScaler use karaddi meka oni wenawa).

## 5. Variance & Standard Deviation - Data Spread eka

Meka podak patalenawa wage hithunata baya wenna epa. Case eka simple. 

### Variance kiyanne mokakda?:
Variance eken balanne ape data points, mean eke idan **kochchara athata wisirila (spread wela) thiyenawada** kiyala.

### Formula eka:
$$\sigma^2 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})^2}{n}$$

Where:
- $x_i$ = hama data point ekakma
- $\bar{x}$ = mean eka
- $n$ = okkoma data gana

### Standard Deviation:
Variance eke **square root (warga mulaya)** gaththama Standard Deviation eka enawa. Meka thamai api godak welawata use karanne mokada unit eka same nisa.

$$\sigma = \sqrt{\sigma^2}$$

```python
# Step by step hadala balamu
data = [4, 8, 6, 5, 3]

# Step 1: Mean eka hoyamu
mean = np.mean(data)
print(f"Data: {data}")
print(f"Mean eka: {mean}")

# Step 2: Mean eken kochchara wenasda? (Difference)
differences = [x - mean for x in data]
print(f"Differences: {differences}")

# Step 3: Ewa square karamu (naththam negative ewa nisa awul)
squared_diff = [d**2 for d in differences]
print(f"Squared differences: {squared_diff}")

# Step 4: Mewage average eka ganna one = Variance
variance = sum(squared_diff) / len(data)
print(f"Variance: {variance}")

# Step 5: Square root karama = Standard Deviation
std_dev = math.sqrt(variance)
print(f"Standard Deviation: {std_dev:.4f}")

print("\nNumpy walin lesiyenma:")
print(f"Variance: {np.var(data)}")
print(f"Std Dev: {np.std(data):.4f}")
```

```python
# Real-life scene ekak: Class dekaka marks
class_a = [70, 72, 71, 73, 69]  # Meyalage marks laga laga yanne (Consistent)
class_b = [50, 90, 60, 85, 70]  # Meyalage marks tikak varied wage (Varied)

print("Class A scores:", class_a)
print(f"   Mean: {np.mean(class_a):.2f}")
print(f"   Std Dev: {np.std(class_a):.2f}")

print("\nClass B scores:", class_b)
print(f"   Mean: {np.mean(class_b):.2f}")
print(f"   Std Dev: {np.std(class_b):.2f}")

print("\n balanna, dekema mean eka same. Habai Class B wala Std Dev eka wadi. E kiyanne eyalage samaharu hodata karala, samaharu anaagena.")
```

### üí° Insight: Standard Deviation
- **Low Std Dev:** Consistent / Predictable.
- **High Std Dev:** Risky / Volatile.
- **ML Use:** Feature Scaling (StandardScaler) ‡∂ß ‡∂Ö‡∂±‡∑í‡∑Ä‡∑è‡∂ª‡∑ä‡∂∫‡∂∫‡∑ô‡∂±‡∑ä ‡∂ï‡∂±.

```python
# üé® Visual: Standard Deviation (Spread)
# Low Std Dev = ‡∑Ö‡∂ü ‡∑Ö‡∂ü (Consistent)
# High Std Dev = ‡∑Ä‡∑í‡∑É‡∑í‡∂ª‡∑í‡∂Ω‡∑è (Volatile/Risky)

data_consistent = np.random.normal(70, 5, 500)  # Marks: Low variation
data_volatile = np.random.normal(70, 20, 500)   # Marks: High variation

plt.figure(figsize=(10, 4))
sns.histplot(data_consistent, color='green', alpha=0.6, label=f'Consistent (Std‚âà5)', kde=True)
sns.histplot(data_volatile, color='red', alpha=0.6, label=f'Volatile (Std‚âà20)', kde=True)
plt.legend()
plt.title('Which class would you trust? (Green = Predictable)')
plt.show()
```

```python
# Visualizing Spread (Standard Deviation)
# Normal (Low Std Dev) vs Spiky (High Std Dev) wage neme.
# Spread eka wadi unama curve eka 'Flat' wenawa.

data_low_std = np.random.normal(50, 5, 1000)   # Std Dev = 5 (Laga laga)
data_high_std = np.random.normal(50, 15, 1000) # Std Dev = 15 (Wisirila)

plt.figure(figsize=(10, 5))
sns.histplot(data_low_std, color='blue', alpha=0.5, label='Low Std Dev (Consistent)', kde=True)
sns.histplot(data_high_std, color='red', alpha=0.5, label='High Std Dev (Risky)', kde=True)
plt.legend()
plt.title('Standard Deviation Graph eken balamu')
plt.show()
```

## 6. Probability - ‡∑É‡∂∏‡∑ä‡∂∑‡∑è‡∑Ä‡∑í‡∂≠‡∑è‡∑Ä (Chance eka)

Probability kiyanne deyak wenna thiyena **chance eka kochcharada** kiyana eka 

### The Formula:
$$P(A) = \frac{\text{Apita oni de wena waragana}}{\text{Okkoma wena waragana}}$$

### Range eka:
- 0 = Wennewama na (Impossible)
- 1 = Aniwa wenawa (Certain - 100%)
- 0.5 = 50-50 chance ekak

Example tikak baluwama therei.

```python
# Dice roll scene eka
print("Dice ekak dammama:")
print("\nTotal outcomes = 6 (E kiyanne 1, 2, 3, 4, 5, 6 enna puluwan)")

# P(4 watena eka)
favorable = 1  # 4 thiyenne ekai ne
total = 6
p_four = favorable / total
print(f"\nP(getting 4) = {favorable}/{total} = {p_four:.4f} = {p_four*100:.2f}%")

# P(Iratte sankyawak watena eka)
favorable_even = 3  # 2, 4, 6
p_even = favorable_even / total
print(f"P(even number) = {favorable_even}/{total} = {p_even:.4f} = {p_even*100:.2f}%")

# P(4 ta wada wadi eka)
favorable_gt4 = 2  # 5, 6
p_gt4 = favorable_gt4 / total
print(f"P(number > 4) = {favorable_gt4}/{total} = {p_gt4:.4f} = {p_gt4*100:.2f}%")
```

```python
# Coin flip simulation!
import random

def flip_coin(n_flips):
    results = [random.choice(['Heads', 'Tails']) for _ in range(n_flips)]
    heads_count = results.count('Heads')
    tails_count = results.count('Tails')
    return heads_count, tails_count

# 10 parak flip karamu
heads, tails = flip_coin(10)
print(f"10 parak flip kara:")
print(f"   Heads: {heads} ({heads/10*100:.1f}%)")
print(f"   Tails: {tails} ({tails/10*100:.1f}%)")

# 1000 parak flip karamu - Dan thamai niyama results enne
heads, tails = flip_coin(1000)
print(f"\n1000 parak flip kara:")
print(f"   Heads: {heads} ({heads/1000*100:.1f}%)")
print(f"   Tails: {tails} ({tails/1000*100:.1f}%)")

print("\nApi thawa thawa flip karanna karanna, value eka 50% paththata lan wenawa. Okata thamai Law of Large Numbers kiyanne.")
```

### üí° Insight: Probability in ML
- ML model ‡∂ë‡∂ö‡∂ö‡∑ä ‡∂ö‡∑í‡∂∫‡∂±‡∑ä‡∂±‡∑ö '‡∂∏‡∑ö‡∂ö ‡∂∂‡∂Ω‡∑ä‡∂Ω‡∑ô‡∂ö‡∑ä' ‡∂±‡∑ô‡∂∏‡∑ô‡∂∫‡∑í.
- '‡∂∏‡∑ö‡∂ö ‡∂∂‡∂Ω‡∑ä‡∂Ω‡∑ô‡∂ö‡∑ä ‡∑Ä‡∑ô‡∂±‡∑ä‡∂± **90% Probability** ‡∂ë‡∂ö‡∂ö‡∑ä ‡∂≠‡∑í‡∂∫‡∑ô‡∂±‡∑Ä‡∑è' ‡∂ö‡∑í‡∂∫‡∂Ω‡∑è.

### Probability Rules podak (Important):

**1. Addition Rule (OR - Meaka hari Araka hari):**
$$P(A \text{ or } B) = P(A) + P(B) - P(A \text{ and } B)$$

**2. Multiplication Rule (AND - Mekai Arakai dekama):**
$$P(A \text{ and } B) = P(A) \times P(B|A)$$
Dekama independent nam (eka matha anika depend wenne nattam), direct guna karanna puluwan.

```python
# Dice deka scene eka
print("Dice dekak damma kiyamu:")

# P(dekema 6 watena eka)
p_six = 1/6
p_both_six = p_six * p_six  # Independent nisa kelinma guna kara
print(f"\nP(dekema 6 watena eka) = (1/6) √ó (1/6) = {p_both_six:.4f}")

# P(aduma gane eka 6k wath watena eka)
p_no_six = 5/6
p_no_six_both = p_no_six * p_no_six
p_at_least_one_six = 1 - p_no_six_both
print(f"P(aduma gane ekak hari 6) = 1 - P(6k naththema na) = 1 - (5/6)¬≤ = {p_at_least_one_six:.4f}")
```

## 7. Percentiles & Quartiles - ‡∂¥‡∑ä‚Äç‡∂ª‡∂≠‡∑í‡∑Å‡∂≠ ‡∑É‡∑Ñ ‡∂ö‡∑è‡∂ª‡∑ä‡∂≠‡∑î

### Percentile kiyanne mokakda?:
Percentile kiyanne **data set eke yam agayakata wada aduwen thiyena percentage eka**.

Example ekak kiwwoth: oyage exams marks 90th percentile nam, e kiyanne panthiye 90%k lamainta wada oya hodata karala.

### Quartiles (Hatarata bedamu):
- **Q1** (25th percentile) - Lower quartile (Palleha kalla)
- **Q2** (50th percentile) - Median (Mada)
- **Q3** (75th percentile) - Upper quartile (Uda kalla)
- **IQR** = Q3 - Q1 (Interquartile Range - Mada 50% range eka)

```python
# Exam scores tikak
scores = [65, 70, 75, 80, 82, 85, 88, 90, 92, 95, 98, 100]

print(f"Exam scores: {scores}")
print(f"\nQuartiles tika:")
print(f"   Q1 (25th percentile): {np.percentile(scores, 25)}")
print(f"   Q2 (50th percentile/Median): {np.percentile(scores, 50)}")
print(f"   Q3 (75th percentile): {np.percentile(scores, 75)}")

q1 = np.percentile(scores, 25)
q3 = np.percentile(scores, 75)
iqr = q3 - q1
print(f"\nIQR = Q3 - Q1 = {q3} - {q1} = {iqr}")

# Uba 90 gaththoth koheda inne?
print(f"\nOyage score eka 90 nam oya inne {scipy_stats.percentileofscore(scores, 90):.1f}th percentile eke")
```

```python
# Box plot data ekak mathak karaganna
data = [12, 15, 18, 22, 25, 28, 30, 35, 40, 45, 50, 80]

q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Outlier boundaries hoyamu
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print(f"Data: {data}")
print(f"\nBox Plot Statistics:")
print(f"   Minimum: {min(data)}")
print(f"   Q1: {q1}")
print(f"   Q2 (Median): {q2}")
print(f"   Q3: {q3}")
print(f"   Maximum: {max(data)}")
print(f"   IQR: {iqr}")
print(f"\nOutlier boundaries: [{lower_bound:.2f}, {upper_bound:.2f}]")
print(f"Ara anthima value eka 80 balanna. Eka outlier ekak (> {upper_bound:.2f}). Eka range eken eliye thiyenne.")
```

```python
# Visualizing the outlier
import matplotlib.pyplot as plt
import seaborn as sns

# Box plot example visualize karamu
data_outlier = [12, 15, 18, 22, 25, 28, 30, 35, 40, 45, 50, 80] # 80 kiyanne outlier eka

plt.figure(figsize=(10, 4))
sns.boxplot(x=data_outlier, color='cyan')
plt.title('Box Plot: Ara 80 outlier eka thaniyama inna hati balanna')
plt.xlabel('Values')
plt.show()
```

## 8. Correlation - ‡∑É‡∑Ñ‡∑É‡∂∏‡∑ä‡∂∂‡∂±‡∑ä‡∂∞‡∂≠‡∑è‡∑Ä‡∂∫

Correlation kiyanne machan **variables dekak athara connection ekak thiyenawada** kiyala balana eka!

### Correlation Coefficient (r) kiyala ekak thiyenawa:
- **r = 1**: Perfect positive (Ekak wadi weddi anikath wadi wenawa - Best friends wage)
- **r = -1**: Perfect negative (Ekak wadi weddi anika adu wenawa - Hapanaya wage)
- **r = 0**: No correlation (Kisima sambandayak na)

### Formula (Just dakala purudu wenna):
$$r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}}$$

```python
# Study hours vs Exam score (Positive scene ekak)
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [40, 50, 55, 65, 70, 78, 85, 92]

correlation = np.corrcoef(study_hours, exam_scores)[0, 1]

print("Study Hours vs Exam Scores:")
print(f"   Study Hours: {study_hours}")
print(f"   Exam Scores: {exam_scores}")
print(f"\nCorrelation coefficient (r): {correlation:.4f}")

if correlation > 0.7:
    print("\nStrong positive correlation!")
    print("   Wadi pura padam karana tharamata marks wadi wenawa. Eka aniwa wena deyak ne.")
```

```python
# Visualizing Correlation
plt.figure(figsize=(8, 6))
sns.scatterplot(x=[20, 22, 25, 28, 30, 32, 35], y=[100, 120, 150, 200, 250, 280, 350], s=100, color='orange')
plt.title('Ice Cream Sales vs Temperature (Positive Correlation)')
plt.xlabel('Temperature (C)')
plt.ylabel('Ice Cream Sales')
plt.show()

print('Pattern eka dakka neda? Line ekak wage udata yanawa.')
```

```python
# Negative correlation example (Ninda vs Stress)
hours_of_sleep = [8, 7, 6, 5, 4, 3, 2, 1]
stress_level = [1, 2, 3, 4, 5, 6, 7, 9]

correlation_neg = np.corrcoef(hours_of_sleep, stress_level)[0, 1]

print("Hours of Sleep vs Stress Level:")
print(f"   Sleep Hours: {hours_of_sleep}")
print(f"   Stress Level: {stress_level}")
print(f"\nCorrelation coefficient (r): {correlation_neg:.4f}")

if correlation_neg < -0.7:
    print("\nStrong negative correlation!")
    print("   Machan nidaganna eka tika adu weddi, Stress eka wadi wenawa Hoda nindak oni apita.")
```

```python
# No correlation example (Sambandayak nathi dewal)
shoe_size = [7, 8, 9, 10, 8, 9, 7, 10]
iq_score = [100, 95, 120, 85, 110, 90, 115, 105]

correlation_none = np.corrcoef(shoe_size, iq_score)[0, 1]

print("Shoe Size vs IQ Score:")
print(f"   Shoe Size: {shoe_size}")
print(f"   IQ Score: {iq_score}")
print(f"\nCorrelation coefficient (r): {correlation_none:.4f}")

if abs(correlation_none) < 0.3:
    print("\nNo significant correlation!")
    print("   Sapaththu size ekai IQ ekai athara kisima connection ekak na. Eka therena deyak ne.")
```

### üí° Insight: Correlation $\neq$ Causation
- High Correlation = Connection ‡∂ë‡∂ö‡∂ö‡∑ä ‡∂≠‡∑í‡∂∂‡∑ä‡∂∂‡∂≠‡∑ä, Reason ‡∂ë‡∂ö ‡∂±‡∑ô‡∂∏‡∑ô‡∂∫‡∑í!
- **ML Use:** Multicollinearity check ‚Üí Feature Selection.
- **Example:** Ice cream sales ‚Üë & Shark attacks ‚Üë (Reason: Summer, not Ice cream!)

```python
# üé® Visual: Correlation (Scatter Plot)
temperature = [20, 25, 28, 30, 32, 35]
ice_cream_sales = [100, 200, 350, 500, 600, 800]

plt.figure(figsize=(6, 4))
plt.scatter(temperature, ice_cream_sales, color='red', s=100)
plt.plot(temperature, ice_cream_sales, 'b--', alpha=0.5) # Trend line
plt.title('Positive Correlation: Temperature ‚Üë Sales ‚Üë')
plt.xlabel('Temperature (¬∞C)')
plt.ylabel('Ice Cream Sales')
plt.show()

print(f'Correlation: {np.corrcoef(temperature, ice_cream_sales)[0,1]:.2f}')
```
